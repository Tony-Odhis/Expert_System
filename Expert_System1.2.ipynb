```python
#the inference engine 
def diagnose_disease(disease_data): #from the knwoledgde base
    print("Welcome to the Plant Disease Diagnosis System!")
    answers = {}  # Store user answers

    for disease, data in disease_data.items():
        print(f"\nChecking for: {disease}")
        answers[disease] = []
        for question in data["questions"]:
            while True:
                user_input = input(f"{question} (yes/no): ").lower()
                if user_input in ("yes", "no"):
                    answers[disease].append(user_input)
                    break
                else:
                    print("Invalid input. Please enter 'yes' or 'no'.")

    # Matching logic:
    possible_diseases = []
    for disease, data in disease_data.items():
        match_count = 0
        for i, answer in enumerate(answers[disease]):
            if answer == "yes" and data["symptoms"][i] is not None: # check if the symptom exist
                match_count += 1
        if match_count > 0:
            possible_diseases.append((disease, match_count))

    if possible_diseases:
        possible_diseases.sort(key=lambda x: x[1], reverse=True) # sort by number of matches
        print("\nPossible Diagnoses:")
        for disease, match_count in possible_diseases:
            print(f"- {disease} (Matched {match_count} symptoms)")
    else:
        print("\nNo matching disease found based on your answers.")

diagnose_disease(disease_data)
```
